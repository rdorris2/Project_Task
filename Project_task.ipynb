```python
# import urllib, json and pandas libraries
from urllib.request import urlopen
import json
import pandas as pd
 
# store the response of url
response = urlopen("https://randomuser.me/api/?results=500")
  
# storing the json response 
data_json = json.loads(response.read())

# create dataframe from json response
df = pd.json_normalize(data_json, record_path=["results"], sep='_')

# display the first 10 records of dataframe
df.head(10)
```
